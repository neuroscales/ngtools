```python
import os
import requests
import fsspec
import  zarr.storage

os.environ["LINCBRAIN_API_KEY"] = "4e4d001a48248d00a8a6e72b9ce808d40607ce57"

LINCBRAIN_API: str = "https://api.lincbrain.org/api"

def linc_auth_opt(token = None):
    """
    Create fsspec authentication options to access lincbrain data.

    Parameters
    ----------
    token : str
        Your LINCBRAIN_API_KEY

    Returns
    -------
    opt : dict
        options to pass to `fsspec`'s `HTTPFileSystem`.
    """
    token = token or os.environ.get('LINCBRAIN_API_KEY', None)
    if not token:
        return {}
    # Check that credential is correct
    session = requests.Session()
    session.get(
        f"{LINCBRAIN_API}/auth/token",
        headers={"Authorization": f"token {token}"}
    )
    # Get cookies
    response = session.get(
        f"{LINCBRAIN_API}/permissions/s3/",
        headers={"Authorization": f"token {token}"}
    )
    cookies = response.cookies.get_dict()
    # Pass cookies to FileSystem
    opt = dict(client_kwargs={'cookies': cookies})
    return opt


def filesystem(protocol, **opt):
    """Return the filesystem corresponding to a protocol or URI."""
    if isinstance(protocol, fsspec.AbstractFileSystem):
        return protocol
    protocol = str(protocol)
    if "://" in protocol:
        protocol = protocol.split("://")[0]
    else:
        protocol = "file"
    try:
        opt.update(linc_auth_opt())
    except Exception:
        pass
    print(opt)
    return fsspec.filesystem(protocol, **opt)


def store(upath, **opt):
    fs = filesystem(upath, **opt)
    return zarr.storage.FSStore(str(upath), fs=fs)
```

```python
def omezarr_to_nii(inp, out, level = -1):
    if isinstance(inp, str):
        group = zarr.open(store(inp))
    else:
        group = inp
    ome = group.attrs["multiscales"][0]

    if isinstance(level, int):
        key = ome["datasets"][level]["path"]
    else:
        key = level

    dat = np.asarray(group[key])

    nifti_axes = ["x", "y", "z", "t", "c"]
    axes = ome["axes"]
    axes_names = [axis["name"] for axis in axes]
    for dataset in ome["datasets"]:
        if dataset["path"] == key:
            break
    time_transforms = ome.get("coordinateTransformations", [])
    space_transforms = dataset["coordinateTransformations"]

    perm = []
    explored_axes = []
    for name in nifti_axes:
        if all(axis["name"] in explored_axes for axis in axes):
            break
        foundit = False
        for i, axis in enumerate(axes):
            if axis["name"] == name:
                perm += [i]
                explored_axes.append(name)
                foundit = True
                break
        if not foundit:
            dat = dat[..., None]
            perm += [dat.ndim-1]

    dat = dat.transpose(perm)
    nifti_axes = nifti_axes[:dat.ndim]

    unit_to_nifti = {
        "nanometer": 1e-6,
        "micrometer": 1e-3,
        "millimeter": 1,
        "meter": 1e3,
        "nanosecond": 1e-9,
        "microsecond": 1e-6,
        "millisecond": 1e-3,
        "second": 1,
    }

    affine = np.eye(4)
    unit = [axis.get("unit", "") for axis in axes]
    for transform in space_transforms:
        if "scale" in transform:
            value = transform["scale"]
        elif "translation" in transform:
            value = transform["translation"]
        value = [
            value[i] * unit_to_nifti.get(unit[i], 1)
            if i < len(value) else 1.0
            for i in perm
        ]
        if "scale" in transform:
            affine[[0, 1, 2], [0, 1, 2]] = value
        else:
            affine[:3, -1] = value


    hdr = nib.Nifti1Header()
    zooms = affine[[0, 1, 2], [0, 1, 2]]

    if "t" in nifti_axes:
        index_time = axes_names.index("t")
        for transform in time_transforms:
            if "scale" in transform:
                value = transform["scale"]
                value = value[index_time] * unit.get(unit[index_time], 1)
                zooms.append(value)

    hdr.set_data_shape(dat.shape)
    hdr.set_xyzt_units("mm", "sec")
    hdr.set_zooms(zooms)

    nib.save(nib.Nifti1Image(dat, affine, hdr), out)
```

```python
colormap_orient = """
vec3 colormapOrient(vec3 orient) {
  vec3 result;
  result.r = abs(orient[0]);
  result.g = abs(orient[1]);
  result.b = abs(orient[2]);
  return clamp(result, 0.0, 1.0);
}
"""

# Runing this in the skeleton shader uses orientation to color tracts
orient_shader = colormap_orient + """
#uicontrol bool orient_color checkbox(default=true)
void main() {
  if (orient_color)
    emitRGB(colormapOrient(orientation));
  else
    emitDefault();
}
"""

def rotate_orient_shader(mat):
    return (colormap_orient +
"""
#uicontrol bool orient_color checkbox(default=true)
void main() {
"""
+
f"""
    mat3 mat = mat3(
      {mat[0, 0]}, {mat[1, 0]}, {mat[2, 0]},
      {mat[0, 1]}, {mat[1, 1]}, {mat[2, 1]},
      {mat[0, 2]}, {mat[1, 2]}, {mat[2, 2]}
    );
    vec3 orient_rot = mat * orientation;
  if (orient_color)
    emitRGB(colormapOrient(orient_rot));
  else
    emitDefault();
"""
+
"""
}
"""
)

ic_shader = """
#uicontrol invlerp normalized
void main() {
  vec4 rgba;
  rgba.r = normalized();
  rgba.a = normalized();
  emitRGBA(rgba);
}
"""
```

```python
ras = ng.CoordinateSpace({"x": [1, "mm"], "y": [1, "mm"], "z": [1, "mm"]})
hipct_01_02 = ng.CoordinateSpaceTransform(
    input_dimensions=ras,
    output_dimensions=ras,
    matrix=XFM["HIPCT_01_02"][:-1]
)
hipct_02_01 = ng.CoordinateSpaceTransform(
    input_dimensions=ras,
    output_dimensions=ras,
    matrix=np.linalg.inv(XFM["HIPCT_01_02"])[:-1]
)
hipct_01_03 = ng.CoordinateSpaceTransform(
    input_dimensions=ras,
    output_dimensions=ras,
    matrix=XFM["HIPCT_01_03"][:-1]
)
hipct_03_01 = ng.CoordinateSpaceTransform(
    input_dimensions=ras,
    output_dimensions=ras,
    matrix=np.linalg.inv(XFM["HIPCT_01_03"])[:-1]
)
dwi_hipct = load_affine(XFM["DWI_HIPCT_init"], names=("x", "y", "z"))
hipct_dwi = inverse(dwi_hipct)
hipct_ras = load_affine(HIPCT2RAS, names=("x", "y", "z"))
hipct_ras = inverse(hipct_ras)


def fix_hipct_space(hipct: ZarrDataSource) -> ZarrDataSource:
    hipct.transform = ng.CoordinateSpaceTransform(
        input_dimensions=ng.CoordinateSpace({
            key: [scale, "um" if unit == "μm" else unit]
            for key, (scale, unit)
            in hipct.transform.input_dimensions.to_json().items()
        }),
        output_dimensions=ng.CoordinateSpace({
            key: [scale, "um" if unit == "μm" else unit]
            for key, (scale, unit)
            in hipct.transform.output_dimensions.to_json().items()
        }),
        matrix=hipct.transform.matrix,
    )
    return hipct


hipct_hemi =  fix_hipct_space(ZarrDataSource("zarr://" + HIPCT["hemi"]))
hipct_hemi.transform = compose(hipct_ras, hipct_hemi.transform)

hipct_broca = fix_hipct_space(ZarrDataSource("zarr://" + HIPCT["broca"]))
hipct_broca.transform = compose(hipct_ras, hipct_01_02, hipct_broca.transform)

hipct_brainstem = fix_hipct_space(ZarrDataSource("zarr://" + HIPCT["brainstem"]))
hipct_brainstem.transform = compose(hipct_ras, hipct_01_03, hipct_brainstem.transform)

dwi_hemi = f"http://127.0.0.1:{port}" + DWI_2HIPCT["hemi"]["b0"]
dwi_hemi = ZarrDataSource(dwi_hemi)
dwi_hemi.transform = compose(hipct_ras, dwi_hemi.transform)

dwi_ic1 = f"http://127.0.0.1:{port}" + DWI_2HIPCT["ic1"]["b0"]
dwi_ic1 = ZarrDataSource(dwi_ic1)
dwi_ic1.transform = compose(hipct_ras, dwi_ic1.transform)

# hipct_hemi.transform = compose(hipct_dwi, hipct_hemi.transform)

def normalize(source):
    mn, q0, q1, mx = source.quantiles([0.0, 0.01, 0.99, 1.0])
    return {
        "normalized": {
            "range": np.stack([q0, q1]).tolist(),
            "window": np.stack([mn, mx]).tolist(),
        }
    }

hipct_hemi = ng.ImageLayer({
    "source": hipct_hemi.to_json(),
    "shaderControls": normalize(hipct_hemi)
})
hipct_broca = ng.ImageLayer({
    "source": hipct_broca.to_json(),
    "shaderControls": normalize(hipct_broca)
})
hipct_brainstem = ng.ImageLayer({
    "source": hipct_brainstem.to_json(),
    "shaderControls": normalize(hipct_brainstem)
})
dwi_hemi = ng.ImageLayer({
    "source": dwi_hemi.to_json(),
    "shaderControls": normalize(dwi_hemi)
})
dwi_ic1 = ng.ImageLayer({
    "source": dwi_ic1.to_json(),
    "shaderControls": normalize(dwi_ic1),
    "shader": ic_shader,
})

with viewer.txn() as state:
    state.layers.append(
        name="HiP-CT hemi",
        layer=hipct_hemi,
    )
    state.layers.append(
        name="HiP-CT broca",
        layer=hipct_broca,
    )
    state.layers.append(
        name="HiP-CT brainstem",
        layer=hipct_brainstem,
    )
    state.layers.append(
        name="DWI (b=0) hemi",
        layer=dwi_hemi,
    )
    state.layers.append(
        name="DWI (b=0) internal capsule",
        layer=dwi_ic1,
    )
    state.crossSectionOrientation = [0, 1, -1, 0]
    print(state.to_json())
```
